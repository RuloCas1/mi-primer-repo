```python
total = range(1,36,1)
```

```python
n=0
url = "https://mx.indeed.com/cmp/Ppg-Comex/reviews?fcountry=ALL"
empresa = 'PPG COMEX'
#data = pd.DataFrame(columns=('id','rate','title','review','cargo','vigente','origen','fecha','pros','contras'))
pros =' '
contras =' '

#evaluaciones = rev.find_all('div','css-ezfk60 eu4oa1w0')[0].text

for p in total:
    urln = url+str(p)
    print(urln)
    driver.get(urln)
    driver.maximize_window()
    desplazar = random.randint(400, 2200)
    driver.execute_script('window.scrollTo(1, '+str(desplazar)+');')
    sleep = random.randint(3, 5)
    time.sleep(sleep)
    elem = driver.find_element_by_xpath("//*")
    soup = BeautifulSoup(elem.get_attribute('innerHTML'),'html.parser')
    reviews = soup.find_all("div",'css-5cqmw8 eu4oa1w0')
    print(len(reviews))

    for rev in reviews:
        rate = rev.find_all(attrs={'itemprop': 'ratingValue'})[0].get('content')
        title = rev.find_all('h2','css-wn2j4x e1tiznh50')[0].text
        autor = rev.find_all('div','css-8a5o2x e1wnkr790')
        if autor:
            fecha = autor[0].text.split(" - ")[2].strip()
            cargo = autor[0].text.split(") - ")[0].split('(')[0]
            origen = autor[0].text.split(") - ")[1].split(' - ')[0] + ', Mexico'
            vigente = autor[0].text.split(") - ")[0].split('(')[1]
        else:
            fecha = ()
        review = rev.find_all('span','css-1cxc9zk e1wnkr790')[0].text.replace('\n',' ').strip()

        if len(rev.find_all('','css-rr5fiy eu4oa1w0')) > 1:
            procon = rev.find_all('','css-rr5fiy eu4oa1w0')[1].text
            if procon[:3] == 'Pro':
                if 'Contras' in procon:
                    pros = procon[len('Pros'):procon.find('Contras')]
                    contras = procon[procon.find('Contras')+len('Contras'):]
                if 'Contras' not in procon:
                    pros = procon[len('Pros'):]
                    contras = 'Ninguno'
            elif procon[:3] == 'Con':
                pros = 'Ninguno'
                contras = procon[len('Contras'):]
            else:
                pros ='Ninguno'
                contras ='Ninguno'
            #print(pros,contras)
        n = n+1
        data = data.append({'id': n,'rate':rate,'title':title,'review':review,'cargo':cargo,
                        'vigente':vigente,'origen':origen,'fecha':fecha,'pros':pros,
                        'contras':contras,'contras':contras,'proyecto':empresa},ignore_index=True)

    sleep = random.randint(15, 29)
    #print(sleep)
    time.sleep(sleep)

    data['contras'] = data['contras'].str.replace('\n',' ')
    data['contras'] = data['contras'].str.replace('\r',' ')
    data['review'] = data['review'].str.replace('\n',' ')
    data['review'] = data['review'].str.replace('\r',' ')
    data['pros'] = data['pros'].str.replace('\n',' ')
    data['pros'] = data['pros'].str.replace('\r',' ')
```
